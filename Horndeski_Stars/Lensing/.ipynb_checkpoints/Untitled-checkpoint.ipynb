```python
# solución analítica en r/rs, con rs el radio de Schw

def ASc(rs):
    
    G = np.sqrt((rs-1)*(rs+3))
    k = np.sqrt((G-rs+3)/(2*G))
    phi = np.arcsin(np.sqrt((1+G-rs)/(3+G-rs)))
    
    theta = -np.pi+4*np.sqrt(rs/G)*(elk(k**2)-elf(phi, k**2))
    
    return theta
```

```python
# solución analítica en r/rs, con rs el radio de Schw

def ASc(rs):
    
    G = np.sqrt((rs-1)*(rs+3))
    k = np.sqrt((G-rs+3)/(2*G))
    phi = np.arcsin(np.sqrt((1+G-rs)/(3+G-rs)))
    
    theta = -np.pi+4*np.sqrt(rs/G)*(elk(k**2)-elf(phi, k**2))
    
    return theta
```
