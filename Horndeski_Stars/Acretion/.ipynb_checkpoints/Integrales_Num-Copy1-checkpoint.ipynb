### Ecuaciones


$$
J_0(x)=-\frac{4\pi L_0}{N(r)\rho(r)}\int_{E_c(r)}^m dE E F_0(E)K_l(a,b)
$$
donde $\rho=r\sin(\theta)$, $N(r)=1-2M/r$, 

$$
E_c(r)=m\frac{r-2M}{\sqrt{r(r-3M)}}, \quad \text{si} \quad 4M\leq r\leq 6M
$$

$$
E_c(r)=m\frac{r+2M}{\sqrt{r(r+6M)}}, \quad \text{si} \quad 6M\leq r
$$

$$
F_0(E)=A(1-\frac{E}{E_0})^{k-3/2}
$$

$$
K_l(a,b)=\frac{1}{l+1}\bigg[\left(\frac{a}{b}-1\right)_{+}^{l+1}-b \bar{K}_l(a,b)\bigg]
$$

$$
a(E,\theta)=\frac{L_c(E)\sin{\theta}}{L_0}
$$

$$
b(E,r)=\frac{L_c(E)}{L_{max}(E,r)}
$$

$$
L_{max}(E,r)=r\sqrt{\frac{E^2}{N(r)}-m^2}
$$

$$
L_c(E)=\frac{4\sqrt{2}M m^3}{\sqrt{36 m^2E^2-8m^4-27E^4+E(9E^2-8m^2)^{3/2}}}
$$

$$
\bar{K}_l(a,b)=\frac{2}{\pi}\sqrt{1-b^2}\int_{1/a}^{1}\frac{d\lambda_z \lambda_z}{\sqrt{1-\lambda_z^2}}\frac{(a\lambda_z-1)_{+}^{l+1}}{1-b^2\lambda_z^2}
$$

```python
def K(E, r, th, l, L0):
    """
    fa -> a, fb ->b, fk->\bar{K}
    """
    a = fa(E, th, L0)
    b = ec.fb(E, r)

    # calculando a/b-1
    arg = a>b #a/b-1
    
    #print('argumento', arg)
    
    if a>b: # arg >= 0:
        kl = (arg**(l+1)-b*fk(a, b, l))/(l+1)
        
    else:
        kl = 0

    return kl
```

```python
def fk(a, b, l):
    """
    x -> lambda_z
    """

    #integ = lambda x: x*(a*x-1)**(l+1)/(np.sqrt(1-x**2)*(1-b**2*x**2))
    
    def integ(x):
        #print(x)
        return x*(a*x-1)**(l+1)/(np.sqrt(1-x**2)*(1-b**2*x**2))
    
    bk = 2*np.sqrt(1-b**2)*quad(integ, 1/a, 1)[0]/np.pi

    return bk
```

```python
def J0(r, th, A, k, l, E0, L0, m):
    """
    """
    coef = -(4*np.pi*L0)/(ec.fN(r)*ec.rho(r, th))

    # integrando
    Emin = ec.Ec(r)
    #arg = lambda E: E*F0(E, A, k, E0)*K(E, r, th, l, L0)
    def arg(E):
        print(E)
        return E*ec.F0(E, A, k, E0)*K(E, r, th, l, L0)
        

    #print(coef)

    integ = quad(arg, Emin, m)[0]
    J0 = coef*integ
    return J0
```
