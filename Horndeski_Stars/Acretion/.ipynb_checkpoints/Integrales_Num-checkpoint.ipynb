### Ecuaciones


$$
J_0(x)=-\frac{4\pi L_0}{N(r)\rho(r)}\int_{E_c(r)}^m dE E F_0(E)K_l(a,b)
$$
donde $\rho=r\sin(\theta)$, $N(r)=1-2M/r$, 

$$
E_c(r)=m\frac{r-2M}{\sqrt{r(r-3M)}}, \quad \text{si} \quad 4M\leq r\leq 6M
$$

$$
E_c(r)=m\frac{r+2M}{\sqrt{r(r+6M)}}, \quad \text{si} \quad 6M\leq r
$$

$$
F_0(E)=A(1-\frac{E}{E_0})^{k-3/2}
$$

$$
K_l(a,b)=\frac{1}{l+1}\bigg[\left(\frac{a}{b}-1\right)_{+}^{l+1}-b \bar{K}_l(a,b)\bigg]
$$

$$
a(E,\theta)=\frac{L_c(E)\sin{\theta}}{L_0}
$$

$$
b(E,r)=\frac{L_c(E)}{L_{max}(E,r)}
$$

$$
L_{max}(E,r)=r\sqrt{\frac{E^2}{N(r)}-m^2}
$$

$$
L_c(E)=\frac{4\sqrt{2}M m^3}{\sqrt{36 m^2E^2-8m^4-27E^4+E(9E^2-8m^2)^{3/2}}}
$$

$$
\bar{K}_l(a,b)=\frac{2}{\pi}\sqrt{1-b^2}\int_{1/a}^{1}\frac{d\lambda_z \lambda_z}{\sqrt{1-\lambda_z^2}}\frac{(a\lambda_z-1)_{+}^{l+1}}{1-b^2\lambda_z^2}
$$
