```python
PS_BS = np.loadtxt('PS_L3_0.dat', unpack=True) 
PS_H =  np.loadtxt('PS_L3_1.dat', unpack=True) 
```

```python
fN = lambda r, M : np.sqrt(1-M/(4*np.pi*r))
dfN = lambda r, M: M/(8*np.pi*np.sqrt(1-M/(4*np.pi*r))*r**2)

def ps(r, M):
    return fN(r, M)-r*dfN(r, M)
```
